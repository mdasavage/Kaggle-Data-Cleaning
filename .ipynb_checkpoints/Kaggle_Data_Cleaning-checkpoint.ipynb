```python
# Pandas for dataset traversal
# re for python regular expressions
# numpy for aggregations
# matplotlib for visualization

import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv = '/home/savage/Downloads/fifa21 raw data v2.csv'
```

```python
# Loading our csv file into memory with no mixed type inference
df = pd.read_csv(csv, low_memory=False)
```

```python
# Sample view of the first 5 rows of data
df.head()
```

```python
df.columns
```

```python
# Rename target columns to be more descriptive or easier to work with
# Keeping name in line with our cleaning objectives
df.rename(columns = {'↓OVA': 'Overall Rating', "Weight": "Weight (kg)",
                     "Height": "Height (cm)", "Value": "Value (Euros)", "Wage": "Wage (Euros)"}, inplace=True)
```

```python
# Remove all columns that are not of interest for this project
df.drop([col for col in df.columns if col not in 
         ['Club','Overall Rating', 'Weight (kg)', 'Height (cm)', 
          'Name', 'Nationality', 'Value (Euros)', 'Wage (Euros)', 'Joined']], axis=1, inplace=True)
```

```python
# Sample view of dataset to clean with only desired columns
df
```

```python
# Question: Are there any data points that have express their height in cm?
# Do they follow the imperical form (ex: 6'5")

# Use regular expressions to see if any rows match that pattern
# Returns slice of boolean values for all rows in dataset (True if match, False if not)
s = df['Height (cm)'].str.contains(r'\d\'(\d)*(\")*')
```

```python
# Grab the indices of rows that match the above pattern
non_cm_heights_idx = [idx for idx in range(len(s)) if s[idx] == True]
```

```python
# Show data that doesn't have height in cm form
# Result: These rows also do not have their Weight in kg.
df.loc[non_cm_heights_idx]
```

```python
# Remove cm and kg suffixes for all rows that have them
df['Height (cm)'] = df['Height (cm)'].map(lambda height: re.sub(r'cm','', height), na_action=None)
df['Weight (kg)'] = df['Weight (kg)'].map(lambda weight: re.sub(r'kg','', weight), na_action=None)
```

```python
# Function to convert imperical height values (Feet'Inches") form to metric (cm)
# Example: 6'2" -> 188

# Returns str representation of value because the column dtype is a str
def to_cm_str(value):
    if re.match(r'^\d+$', value):
        return int(value)
    height_components = value[::2]
    feet_in_inches = int(height_components[0]) * 12 + int(height_components[1])
    return int(round(feet_in_inches * 2.54))
    
```

```python
# Similiar to above, but for converting imperical pounds (lbs) to metric (kg)
# Example: 183lbs -> 83
def to_kg_str(value):
    if re.match(r'^\d+$',value):
        return int(value)
    return int(round(int(value[:-3]) * 0.45359237))
    
```

```python
# Apply appropriate function to associated column using map()
df['Height (cm)'] = df['Height (cm)'].map(to_cm_str, na_action=None)
df['Weight (kg)'] = df['Weight (kg)'].map(to_kg_str, na_action=None)
```

```python
# View of previous dataset with converted values
df.loc[non_cm_heights_idx]
```

```python
# 'Club' column values have unnescessary newline chars in front of them
df.head()['Club']
```

```python
# Remove multiple newline chars from 'Club' column values with lambda function and regular expression substitution
df['Club'] = df['Club'].map(lambda club: re.sub(r'(\n)+','', club), na_action=None)
```

```python
# After newline char removal
df.head()['Club']
```

```python
df
```

```python
# 'Wage' and 'Value' column values have Euro sign (€) prefix and Million/Thousand suffixes (M/K)
# For analysis, we want to turn these columns into numerical columns

# Use slices to extract portion of string representing a number
# Multiple that number by 1,000 if suffix is K or 1,000,000 if suffix is M
def make_numerical(value):
    numerical = value[1:-1]
    if numerical == '':
        return 0
    suffix = value[-1]
    if suffix == 'M':
        number = float(numerical) * 1000000
    else:
        number = float(numerical) * 1000
    return number
```

```python
make_numerical(df.loc[0, 'Value (Euros)'])
```

```python
# Apply function to both columns

df['Wage (Euros)'] = df['Wage (Euros)'].map(make_numerical, na_action=None)
df['Value (Euros)'] = df['Value (Euros)'].map(make_numerical, na_action=None)
```

```python
# Verify numerical 'Value' and 'Wage' conversion
df
```

```python
# Convert cleaned and processed column datatypes to their appropriate ones for analysis
#
# 'Height (cm) and Weight (kg)': str -> int
# 'Value (Euros) and Wage (Euros)': str -> float
df.astype({'Height (cm)': 'int', 'Weight (kg)': 'int', 'Value (Euros)': 'float64', 'Wage (Euros)': 'float64' })
```

```python
# Question 1: Club with highest average rating?
# Group all players by club
clubs = df.groupby('Club')
```

```python
# Compute average 'Overall Rating' for each group and sort by descending
average_club_rating = clubs['Overall Rating'].agg(np.average).sort_values(ascending=False)
```

```python
average_club_rating
```

```python
# Question 2: Most valuable nationality (in Euros)?

# Group each player by nationality
nationalities = df.groupby('Nationality')
```

```python
# Calculate sum of each nationality group by 'Value (Euros)' and sort in descending order
nationalities_by_value = nationalities['Value (Euros)'].agg(np.sum).sort_values(ascending=False)
```

```python
nationalities_by_value
```

```python
# Create horizontal bar graph with each bar's 'Value (Euros)' annotated
# Only graph top 10 nationalities
bar_graph = nationalities_by_value[:11].plot.barh(logx=True, figsize=(20,20))
bar_graph.bar_label(bar_graph.containers[0])
```
