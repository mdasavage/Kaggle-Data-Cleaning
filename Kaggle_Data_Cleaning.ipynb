```python
def make_numerical(value):
    numerical = value[1:-1]
    if numerical == '':
        return 0
    suffix = value[-1]
    if suffix == 'M':
        number = float(numerical) * 1000000
    else:
        number = float(numerical) * 1000
    return number
```

```python
make_numerical(df.loc[0, 'Value (Euros)'])
```

```python
df['Value (Euros)'] = df['Value (Euros)'].map(make_numerical, na_action=None)
```

```python
df['Wage (Euros)'] = df['Wage (Euros)'].map(make_numerical, na_action=None)
```

```python
def to_cm_str(value):
    if re.match(r'^\d+$', value):
        return value
    height_components = value[::2]
    feet_in_inches = int(height_components[0]) * 12 + int(height_components[1])
    return str(round(feet_in_inches * 2.54))
    
```

```python
def to_kg_str(value):
    if re.match(r'^\d+$',value):
        return value
    return str(round(int(value[:-3]) * 0.45359237))
    
```

```python
to_cm_str(df.loc[non_cm_heights_idx[0], "Height (cm)"])
```

```python
df['Height (cm)'] = df['Height (cm)'].map(to_cm_str, na_action=None)
df['Weight (kg)'] = df['Weight (kg)'].map(to_kg_str, na_action=None)
```
